```python
#2.3
import cv2 
import numpy as np

#----------------蓝色通道值------------------
blue = np.zeros((300,300,3), dtype=np.uint8)

blue[:,:,0] = 255
print("blue=\n",blue)
cv2.imshow("blue",blue)

#---------------绿色通道值---------------------
green = np.zeros((300,300,3), dtype=np.uint8)
green[:,:,1] = 255
print('green=\n',green)
cv2.imshow('green',green)

#--------------红色通道值---------------------
red = np.zeros((300,300,3),dtype = np.uint8)
red[:,:,2] = 255
print("red=\n",red)
cv2.imshow("red",red)

#-------------释放窗口-----------------------
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#2.6
import cv2
img =cv2.imread('lena.jpg')
cv2.imshow('before',img)
print('访问img[0,0]=',img[0,0])
print('访问img[0,0,0]=',img[0,0,0])
print('访问img[0,0,1]=',img[0,0,1])
print('访问img[0,0,2]=',img[0,0,2])
print('visit img[50,0]=',img[50,0])
print('visit img[100,0]=',img[100,0])

#region1
for i in range(0,50):
    for j in range(0,100):
        for k in range(0,3):
            img[i,j,k] = 255

#region2
for i in range(50,100):
    for j in range(0,100):
        for k in range(0,3):
            img[i,j,k] = 128
#region3
for i in range(100,150):
    for j in range(0,100):
        for k in range(0,3):
            img[i,j,k] = 0
cv2.imshow('after',img)

print('访问img[0,0]=',img[0,0])
print('访问img[0,0,0]=',img[0,0,0])
print('访问img[0,0,1]=',img[0,0,1])
print('访问img[0,0,2]=',img[0,0,2])
print('visit img[50,0]=',img[50,0])
print('visit img[100,0]=',img[100,0])

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#2.16
import cv2
lena = cv2.imread('lena.jpg')
cv2.imshow('lena',lena)
b = lena[:,:,0]
g = lena[:,:,1]
r = lena[:,:,2]
cv2.imshow('b',b)
cv2.imshow('g',g)
cv2.imshow('r',r)
lena[:,:,0] = 0
cv2.imshow('b0',lena)
lena[:,:,1] = 0
cv2.imshow('b0g0',lena)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#2.17
import cv2
lena = cv2.imread('lena.jpg')
b = cv2.split(lena)[0]
g = cv2.split(lena)[1]
r = cv2.split(lena)[2]
cv2.imshow('b',b)
cv2.imshow('g',g)
cv2.imshow('r',r)
b,g,r = cv2.split(lena)
cv2.imshow('B',b)
cv2.imshow('G',g)
cv2.imshow('R',r)

bgr = cv2.merge([b,g,r])
cv2.imshow('bgr',bgr)
rgb = cv2.merge([r,g,b])
cv2.imshow('rgb',rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
